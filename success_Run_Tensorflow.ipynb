```
import json

with open('mapped_commands_full.json') as json_file:
    data = json.load(json_file)
    tf = data['tf']
    torch = data['torch']
    # print("tf:", data['tf'])
```

```
import numpy as np
tensorflow = []
```

```
for key, value in tf.items():
        command_dict = value[0]
        attrs = ".".join(command_dict['attrs'])
        # translates_to_tf = "tf" in command_dict
        translates_to_torch = "torch" in command_dict
        translates_to_tfjs = "tfjs" in command_dict
        tensorflow.append((attrs, translates_to_torch, translates_to_tfjs))
comm = np.vstack(tensorflow)
        # tensorflow.append(attrs, translates_to_torch, translates_to_tfjs)
        # print(attrs)
        # for attributes in command_dict['attrs']:
        #   print(attributes.join("."))
```

```
comm
```

```
comm = np.where(comm=='False', '✘', comm)
comm = np.where(comm=='True', '✔', comm)
comm
```

```
from tabulate import tabulate

print('|                                                |PyTorch|Tensorflow.js|')
print(tabulate(comm, tablefmt='pipe'))
```

```
from pytablewriter import MarkdownTableWriter


writer = MarkdownTableWriter(
        table_name="example_table",
        headers=["", "float", "str"],
        comm,)
writer.write_table()
```
